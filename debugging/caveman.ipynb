```python
# Tracing
class Lab():

    def __init__(self, measurements, researcher, password):
        self.measurements = measurements
        self.researcher = researcher
        self.password = password

    def average(self):
        if len(self.measurements) > 0:
            return sum(self.measurements) / len(self.measurements)
    
    def __str__(self):
        return f"Use __str__ to display measurements: {self.measurements}"
    
    def __repr__(self):
        class_name = type(self).__name__
        return (f"{class_name}("
                f"measurements={self.measurements!r}, "
                f"researcher={self.researcher!r}, "
                f"password={self.password!r})")

day_1 = Lab([0.2, 0.1],'John Doe', 'secret')

commands = [
    ("str(day_1)", str(day_1)),
    ("repr(day_1)", repr(day_1)),
    ("format(day_1)", format(day_1)),
    ('f"{day_1}"', f"{day_1}"),
    ('f"{day_1!s}"', f"{day_1!s}"),
    ('f"{day_1!r}"', f"{day_1!r}"),
    ('f"{day_1 = }"', f"{day_1 = }"),
    ('f"{day_1 = !s}"', f"{day_1 = !s}"),
    ('print(day_1)', print(day_1)),
    ('type(eval(repr(day_1))))', type(eval(repr(day_1)))),
]

RED = '\033[91m'
RESET = '\033[0m'  

for command, output in commands:
    print(f'{command.ljust(25)} {RED}->{RESET} {output}')

# print(day_1)
```
